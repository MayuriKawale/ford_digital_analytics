## Task 3 - Data Cleaning
### Step 1 - Unpack JSON Columns

```python
import json

json_cols = ["device", "geoNetwork", "totals", "trafficSource"]

for col in json_cols:
    df[col] = df[col].apply(json.loads) # converts json string to a python dictionary
    normalized = pd.json_normalize(df[col])  # expands the dictionary into separate columns of a new dataframe
    normalized.columns = [f'{col}_{c}' for c in normalized.columns] # rename the columns with original column info
    df = df.drop(col, axis=1).join(normalized) # drops json columns and adds new separate columns

print(f'New shape: {df.shape}')
df.head()
```

```python
# look for column names, data types and any missing values:
df.info()
```

### Step 2 - Check for missing values

```python
missing = df.isnull().sum() # looks for missing values and sums along the column
missing[missing>0] # all columns with missing values
```

### Step 3 - Fill the missing values

##### `totals_transactionRevenue` is the most important feature. It helps to determine the conversion rate. The missing value implies the number of visits that did not generate any revenue. So, these missing values can be easily replaced by `0`  
##### *bounce* means the visitor left after viewing only one page. The visitor did not bounce imply they visited multiple pages. Missing value implies `0` bounces. Replace missing value in `totals_bounces` with `0`.  
##### `totals_newVisits` refer to number of new vistor that is visiting for the first time. Missing value implies they are not returning visitors. So, replace the missing value with `0`.
##### `trafficSource_keyword` refers to visits using specific keyword via a search engine. Missing value here or other `trafficSource` variables indicate another method was used. So, replace the missing values with `NA` or `not applicable`.
##### `totals_pageviews` has only 100 missing values out of 900K+ values. These can be safely replaced with `0`.

```python
# fill missing value with appropriate values
df['totals_transactionRevenue'] = df['totals_transactionRevenue'].fillna(0) # replace with 0
df['totals_bounces'] = df['totals_bounces'].fillna(0) # replace with 0
df['totals_newVisits'] = df['totals_newVisits'].fillna(0) # replace with 0
df['totals_pageviews'] = df['totals_pageviews'].fillna(0) # replace with 0

sources = ['keyword', 'isTrueDirect', 'referralPath', 'adwordsClickInfo.page', 'adwordsClickInfo.slot', 
           'adwordsClickInfo.gclId', 'adwordsClickInfo.adNetworkType', 'adwordsClickInfo.isVideoAd', 'adContent', 
           'campaignCode']

for source in sources:
    df[f'trafficSource_{source}'] = df[f'trafficSource_{source}'].fillna("not applicable") # replace with 'not applicable'
```

```python
#convert the data type
df['date'] = pd.to_datetime(df['date'], format = '%Y%m%d') # date format
df['visitStartTime'] = pd.to_datetime(df['visitStartTime'], unit = 's') # time stamp in seconds
df['visitId'] = df['visitId'].astype(str) # convert to string

df['fullVisitorId'] = df['fullVisitorId'].astype(str) # convert to string

#convert totals columns to int
cols = ['visits', 'hits', 'pageviews', 'bounces', 'newVisits']
for col in cols:
    df[f'totals_{col}'] = df[f'totals_{col}'].astype(int)

df['totals_transactionRevenue'] = df['totals_transactionRevenue'].astype(float) #convert to float
```
